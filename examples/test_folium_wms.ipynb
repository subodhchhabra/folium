```python
m = folium.Map(location=[44, -73], zoom_start=3)

folium.WmsTileLayer(
    "http://gis.srh.noaa.gov/arcgis/services/NDFDTemps/MapServer/WMSServer",
    format="image/png",
    layers=16,
    ).add_to(m)


folium.TileLayer(
    tiles='http://hfradar.ndbc.noaa.gov/tilesavg.php?s=20&e=100&x={x}&y={y}&z={z}&t=2016-01-20%2012:00:00&rez=1',
    attr='hfradar.ndbc.noaa.gov',
    name='hfradar 1km',
    overlay=True
    ).add_to(m)

folium.LayerControl().add_to(m)

m
```
